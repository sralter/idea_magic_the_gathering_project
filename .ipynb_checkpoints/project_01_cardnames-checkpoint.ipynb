After repeated attempts of applying SQL to the `df_c` dataframe, I'm wondering if I need to get dummy variables for some columns. I'm wondering if `colorIdentity` and `colors` are throwing things off due to the square brackets.

I will work on cleaning up some columns and creating dummy columns. `df_c` will be a clean, non-dummy dataframe, and `df_d` will be for the dummy column version of the dataframe.

````python
# df_c is for non-dummy columns

df_c=df[['name','colorIdentity','colors','manaCost','convertedManaCost','types','subtypes']].reset_index().drop('index',axis=1)

# Rename some columns to make it easier on me:
df_c=df_c.rename(columns={'convertedManaCost':'mv'})

df_c.head(3)
````

````python
# redefine df_c as df_d to have dummy columns

df_d=df[['name','colorIdentity','colors','manaCost','convertedManaCost','types','subtypes']].reset_index().drop('index',axis=1)

# Rename some columns to make it easier on me:
df_d=df_d.rename(columns={'convertedManaCost':'mv'})

df_d.head()
````

````python
# dummies for colorIdentity

dummies=df_d['colorIdentity'].str.join(',').str.get_dummies(',')
df_d=pd.concat([df_d,dummies],axis=1)

# re-order columns
correct_order=['name','colorIdentity','colors','manaCost','mv','types','subtypes','W','U','B','R','G']
df_d=df_d[correct_order]

# rename columns to signify they're from colorIdentity
df_d=df_d.rename(columns={'W':'Wi','U':'Ui','B':'Bi','R':'Ri','G':'Gi'})

df_d.head(3)
````

````python
# dummies for colors

dummies=df_d['colors'].str.join(',').str.get_dummies(',')
df_d=pd.concat([df_d,dummies],axis=1)

# re-order columns
correct_order=['name','colorIdentity','colors','manaCost','mv','types','subtypes','Wi','Ui','Bi','Ri','Gi','W','U','B','R','G']
df_d=df_d[correct_order]

# rename columns to signify they're from colors
df_d=df_d.rename(columns={'W':'Wc','U':'Uc','B':'Bc','R':'Rc','G':'Gc'})

df_d.head(3)
````

````python
# re-order columns
correct_order=['name','manaCost','mv','types','subtypes','colorIdentity','Wi','Ui','Bi','Ri','Gi','colors','Wc','Uc','Bc','Rc','Gc']
df_d=df_d[correct_order]

df_d.head(3)
````

````python
# clean `colorIdentity` column

def clean_text(text):
    return ''.join(filter(str.isalpha, text))

df_d['colorIdentity']=df_d['colorIdentity'].apply(clean_text)

df_d.head(3)
````

````python
# clean `colors` column

df_d['colors']=df_d['colors'].apply(clean_text)
df_d.head(3)
````

````python
# clean `manaCost` column
import re

def filter_alphanumeric(text):
    if text is not None and isinstance(text, str):
        return ''.join(re.findall(r'[a-zA-Z0-9]', text))
    else:
        return ''  # Return empty string for 
                   # missing or non-string values
    
df_d['manaCost']=df_d['manaCost'].apply(filter_alphanumeric)
df_d.head(3)
````

````python
# pull out generic mana costs from the `manaCost` column
# to make a new column

def filter_numeric(text):
    if text is not None and isinstance(text, str):
        return ''.join(re.findall(r'[0-9]', text))
    else:
        return ''
    
# gmc for generic mana cost within the mana cost
# to just show the numbers within the mana cost
df_d['gencolorcost']=df_d['manaCost'].apply(filter_numeric)

# re-order columns
correct_order=['name','manaCost','gencolorcost','mv','types','subtypes','colorIdentity','Wi','Ui','Bi','Ri','Gi','colors','Wc','Uc','Bc','Rc','Gc']
df_d=df_d[correct_order]

df_d.head(3)
````

````python
# create column of just the manaCost letters
# without the numbers

def remove_nums(text):
    return ''.join(filter(str.isalpha,text))
    
df_d['colorcost']=df_d['manaCost'].apply(remove_nums)

# re-order columns
correct_order=['name','manaCost','colorcost','gencolorcost','mv','types','subtypes','colorIdentity','Wi','Ui','Bi','Ri','Gi','colors','Wc','Uc','Bc','Rc','Gc']
df_d=df_d[correct_order]

df_d.head(3)
````

````python
# add dummies for types and subtypes

# types
dummies=df_d['types'].str.join(',').str.get_dummies(',')
df_d=pd.concat([df_d,dummies],axis=1)

# subtypes
dummies=df_d['subtypes'].str.join(',').str.get_dummies(',')
df_d=pd.concat([df_d,dummies],axis=1)

df_d.head(3)
````
